```python
netflix_dataset['director'].fillna('Unknown', inplace=True)
netflix_dataset['country'].fillna('Unknown', inplace=True)
```

```python
netflix_dataset.dropna(subset=['title'], inplace=True)
```

```python
netflix_dataset.duplicated().sum()
```

```python
netflix_dataset['country'] = netflix_dataset['country'].str.strip().str.title()
```

```python
netflix_dataset['type'] = netflix_dataset['type'].str.strip().str.title()
```

```python
netflix_dataset['date_added'] = pd.to_datetime(netflix_dataset['date_added']).dt.strftime('%d-%m-%y')
```
